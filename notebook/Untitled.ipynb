```python
import pandas as pd
```

```python
features = pd.read_csv('../data/raw/features.csv')
stores = pd.read_csv('../data/raw/stores.csv')
train = pd.read_csv('../data/raw/train.csv')
test = pd.read_csv('../data/raw/test.csv')
sub = pd.read_csv('../data/raw/sampleSubmission.csv')
```

```python
feature_store = pd.merge(features, stores, how='inner', on = "Store")
train_df = pd.merge(train, feature_store, how='inner', on = ['Store','Date','IsHoliday']).sort_values(by = ['Store','Dept','Date']).reset_index(drop=True)
test_df  = pd.merge(test, feature_store, how='inner', on = ['Store','Date','IsHoliday']).sort_values(by = ['Store','Dept','Date']).reset_index(drop=True)
```
